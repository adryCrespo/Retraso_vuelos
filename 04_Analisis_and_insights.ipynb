# Conclusiones estudio

Este analisis cuenta con datos de enero de 2019 y 2020. El objeto del estudio es intentar ver que factores estan dentras del retraso de los vuelos.  

La probabilidad de retraso por vuelo es del 15%

**INSIGHT 1**:   


Hasta la tercera semana el retraso de los vuelos es prácticamente igual. La diferencia esta en la 4 y 5 (19 al 26)

```python
df['semana_year_salida'] = df['salida_date'].dt.isocalendar().week
aux = df.groupby(['semana_year_salida','year']).dep_del15.mean()

fig, ax = plt.subplots()
aux.unstack(level=1).plot.line(ax=ax)
plt.xticks(np.arange(1, 6, 1))
plt.grid( linestyle=':', linewidth=2)
ax.set_ylabel('ratio retraso');
```

```python
aux = df.groupby(['dia_salida_date','year' ]).dep_del15.mean()
fig, ax = plt.subplots()
g = aux.unstack(level=-1).plot(subplots=False, sharex=False,figsize=(14,7),sharey=True ,ax=ax)
plt.xticks(np.arange(1, 32, 1))
ax.grid( linestyle=':', linewidth=2);
ax.set_ylabel('ratio retraso');
```

**INSIGHT 2**    
Los retrasos va aumentando según aumenta el dia. La probabilidad de retraso aumenta por aeropuerto hasta el 64%

```python
aux = df.groupby(['hora_salida_date','year' ]).dep_del15.mean()

fig, ax = plt.subplots()
aux.unstack(level=1).plot(ax=ax)
plt.xticks(np.arange(0, 26, 2))
ax.grid( linestyle=':', linewidth=2)
ax.set_ylabel('ratio retraso')
plt.show()
```

```python
aux = df.copy()

#Lista de condiciones
condiciones = [(df['hora_salida_date'] < 5),
               (df['hora_salida_date'] >= 5) & (df['hora_salida_date'] < 20 ),
               (df['hora_salida_date'] >= 20)]

#Lista de resultados
resultados = ['madrugada','mediodia','noche']
                
#Aplicamos select
aux['hora_Disc'] = np.select(condiciones,resultados, default = -999)

aux2 = aux.pivot_table(index = 'origin_airport_id', columns = 'hora_Disc', values = 'dep_del15', aggfunc = 'mean', fill_value=-1)
for v in aux2.columns:
    medias = aux2.loc[aux2[v] > 0,v].mean()
    print("ratio retraso %s: %.2f"%(v,medias))
```

**INSIGHT 3**  
cuanto menor sea la frecuencia entre dos aeropuertos mayor es el retraso. Las combinaciones de aeropuertos salida-llega con frecuencias menores a 10 tienen una probabilidad de retraso de un 7% mas que la media.

```python
aux = df[[ 'origin_airport_id', 'dest_airport_id','dep_del15']]\
        .groupby([ 'origin_airport_id', 'dest_airport_id'])[['dep_del15']]\
        .agg(media_retraso=('dep_del15','mean'),conteo=('dep_del15','count')) 


cortes = [0,10,100,500,3500]
#Discretizamos
aux['conteo_disc'] = pd.cut(aux['conteo'],cortes)
fig, ax = plt.subplots()
aux.groupby(['conteo_disc']).media_retraso.mean().sort_index().plot.bar(grid=True,ax=ax)
ax.set_ylabel('retraso medio')
ax.set_xlabel('numero de vuelos entre aeropuertos')
;
```

**INSIGHT 4**  
Los aeropuertos de una misma region tienen tasas de retraso muy distintas

```python
aux = df.groupby(['origin_airport_id','iso_region_org'])['dep_del15'].mean()
aux = aux.reset_index()
aux['aeropuertos_Ranking'] = aux['dep_del15'].rank(method = 'first', ascending = False)
aux.sort_values('dep_del15')
aux2 = aux.groupby('iso_region_org')['aeropuertos_Ranking'].agg(['mean','std']).reset_index()

sns.boxplot(data=aux,x='iso_region_org',y='aeropuertos_Ranking');
```

## Distribucion de los retrasos
- Los aeropuertos con mas vuelos retrasados en absoluto no tienen un ratio de retraso mayor
- los aeropuertos con menos vuelos son los que mas tardan

```python
def pareto_ds4b(variable, salida = 'tabla'):
    #Ordenar en descendente y pasar a dataframe
    pareto = variable.sort_values(ascending = False).to_frame()
    #Cambiar el nombre a la variable
    pareto.columns = ['Valor']
    #Crear la posición
    pareto['Posicion'] = np.arange(start = 1, stop = len(pareto) + 1)
    pareto['Posicion_Porc'] = pareto.Posicion.transform(lambda x: x / pareto.shape[0] * 100)
    #Crear el acumulado
    pareto['Acum'] = pareto['Valor'].cumsum()
    max_pareto_acum = max(pareto.Acum)
    pareto['Acum_Porc'] = pareto.Acum.transform(lambda x: x / max_pareto_acum * 100)
    #Simplificar
    pareto = pareto[['Posicion_Porc','Acum_Porc']]
    
    #Devolver la salida
    if salida == 'grafico':
        f, ax = plt.subplots(figsize = (16,8))
        ax.plot(pareto.Posicion_Porc, pareto.Acum_Porc)
        ax.plot(pareto.Posicion_Porc, pareto.Posicion_Porc)
        ax.tick_params(axis='x', labelsize=12, labelrotation=90)
        return(ax)
    else:
        return(pareto)
```

```python
aux = df.copy()

aux2 = aux.groupby('origin_airport_id')['dep_del15'].sum()
aux3 = pareto_ds4b(variable=aux2)
pareto_ds4b(variable=aux2,salida = 'grafico');
```
